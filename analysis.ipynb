```python
# ===================================
# Board Game Dataset Analysis in Colab
# ===================================

# 1. Install & Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 2. Upload Dataset
from google.colab import files
uploaded = files.upload()

# Load dataset (make sure the uploaded file is named 'lab.csv')
df = pd.read_csv("lab.csv")

# Create a folder for saving plots
if not os.path.exists("plots"):
    os.makedirs("plots")

# Preview dataset
print("Dataset Preview:")
print(df.head())

# -------------------------------
# 3. Data Analysis & Visualization
# -------------------------------

# (a) Number of games released per year
plt.figure(figsize=(10,6))
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title("Number of Board Games Released per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.savefig("plots/games_per_year.png")
plt.show()

# (b) Distribution of average ratings
plt.figure(figsize=(8,6))
sns.histplot(df['avg_rating'], bins=30, kde=True)
plt.title("Distribution of Average Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Frequency")
plt.savefig("plots/rating_distribution.png")
plt.show()

# (c) Min vs Max Players
plt.figure(figsize=(8,6))
sns.scatterplot(x="min_players", y="max_players", data=df, alpha=0.5)
plt.title("Min Players vs Max Players")
plt.xlabel("Min Players")
plt.ylabel("Max Players")
plt.savefig("plots/players_scatter.png")
plt.show()

# (d) Min vs Max Playtime
plt.figure(figsize=(8,6))
sns.scatterplot(x="min_playtime", y="max_playtime", data=df, alpha=0.5)
plt.title("Min vs Max Playtime")
plt.xlabel("Min Playtime (minutes)")
plt.ylabel("Max Playtime (minutes)")
plt.savefig("plots/playtime_scatter.png")
plt.show()

# (e) Top 10 games by average rating
top10 = df.nlargest(10, "avg_rating")[["boardgame", "avg_rating"]]
plt.figure(figsize=(10,6))
sns.barplot(x="avg_rating", y="boardgame", data=top10, palette="viridis")
plt.title("Top 10 Board Games by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Game")
plt.savefig("plots/top10_games.png")
plt.show()

# (f) Ratings vs Number of Ratings
plt.figure(figsize=(8,6))
sns.scatterplot(x="avg_rating", y="num_ratings", data=df, alpha=0.5)
plt.title("Average Rating vs Number of Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Number of Ratings")
plt.savefig("plots/ratings_vs_numratings.png")
plt.show()

print(" All plots have been saved in the 'plots' folder.")
```
